```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from urllib.request import Request,urlopen
```

```python
page_url = requests.get("https://www.passportindex.org/byRank.php")
web_page = BeautifulSoup(page_url.content,'html.parser')
print(web_page)
```

```python
country_name_list = [i for i in web_page.find_all(class_='name_country')]
print(country_name_list)
```

```python
country_name = list()
for i in range(len(country_name_list)):
    country_name.append(country_name_list[i].get_text())
print(country_name)
```

```python
country_rank_list = [r for r in web_page.find_all(class_='name_rank')]
print(country_rank_list)
```

```python
country_rank = list()
for x in range(len(country_rank_list)):
    country_rank.append(country_rank_list[x].get_text())
print(country_rank)
```

```python
visa_free_country = [i for i in web_page.find_all(class_='rank vf')]
#print(visa_free_country)
country_vff = list()
for i in range(len(visa_free_country)):
    country_vff.append(visa_free_country[i].get_text())
print(country_vff)
```

```python
visa_on_arrival = [i for i in web_page.find_all(class_='rank voa')]
#print(visa_on_arrival)
visa_or = list()
for i in range(len(visa_on_arrival)):
    visa_or.append(visa_on_arrival[i].get_text())
print(visa_or)
```

```python
visa_required = [i for i in web_page.find_all(class_='rank vr')]
#print(visa_required)
visa_vr = list()
for i in range(len(visa_required)):
    visa_vr.append(visa_required[i].get_text())
print(visa_vr)
```

```python
dict1={'country':country_name,'Country_Rank':country_rank,'visa_free_country':country_vff,'Visa_On_Arrival':visa_or,'Visa_Required':visa_vr}
```

```python
print(dict1)
```

```python
Passport_Data = pd.DataFrame(
    {
        'country':country_name,
        'Country_Rank':country_rank,
        'visa_free_country':country_vff,
        'Visa_On_Arrival':visa_or,
        'Visa_Required':visa_vr,
    })
```

```python
print(Passport_Data)
```

```python
df = pd.DataFrame.from_dict(dict1)
print(df)
```
